# Case Study 2
In this case study, I receive a dataset from FracFocus and perform an analysis on it. The dataset is very large and many of the attributes that come with the dataset do not contribute to my analysis, so I clean the data before performing any analysis. I focus my analysis on the total water and nonwater susbtances used in the base of the hydraulic system of each well. As water is a key resource in fracking and drilling, it is important to keep track of how much water is needed and whether or not its use causes a negative consequence to the local community's fresh water supply. I group together different counties and analyze how much water is used by the different wells in that county.

# Importing Data from FracFocus
There are multiple large files that will be included in the dataset. I use Pandas DataFrame to store all the data. The DataFrame allows easy manipulation of the data and analysis. I import the `glob` and `os` libraries to help with the multiple files.
<br>**Note:** Because there are so many files, it takes a while for the concatenation part to finish, so please be patient :)

```python
import pandas as pd
import glob
import os

path = r'FracFocusCSV/'
all_files = glob.glob(os.path.join(path, "FracFocusRegistry_*.csv"))
print(f'Here are all the file names that will be used for the DataFrame: {all_files}')
df_from_each_file = (pd.read_csv(f,low_memory=False) for f in all_files)
print(f'Done with getting all the file paths')
concat_df = pd.concat(df_from_each_file, ignore_index=True)
print('Done with concatenating')

# sanity check to make sure all the data was properly stored in the dataframe
print(f'There are {len(concat_df.index)} rows') # how many rows there are: 5667500 rows :O
concat_df.head(10)
```

# Data Cleaning and Processing 
Now that the data has been successfully stored in the DataFrame, I start to clean the data and process it for analysis. First, to get some basic insights of the data, I looked through the data types that were assigned to each value in the data set. I then use the `describe` method to get a statistical summary of the data so that I can use that in my analysis.
<br>**Note:** In Pandas, `object` can be thought of as a Python String. 
## Cleaning
There were a few attributes that were removed as they did not contribute to the analysis of the data. These are:
1. UploadKey
2. IngredientKey
3. IngredientComment 
4. ClaimantCompany
5. DisclosureKey
6. APINumber
7. StateNumber

9. CASNumber
10. SystemApproach

Removing these attributes helped execute the analysis quicker as there were fewer columns to go through. Many of these attributes did not have any values to them, so dropping them was the only option. Other attributes, such as `APINumber` and `StateNumber`, were redundant as there was another attribute that also expressed the same data. <br>
After removing the unneccessary columns, I remove rows that had no value in the `TotalBaseWaterVolume` and `TotalBaseNonWaterVolume` column as these are the columns I am focusing my analysis on.<br>
Finally, I remove the duplicates of the data so that there is only one entry for each well. I did this because the `TotalBaseWaterVolume` and `TotalBaseNonWaterVolume` values appeared multiple times for the same well in the data set.<br>
After cleaning up the data, there are 145734 rows left of the original data set. 
## Data Normalization
Normalizing the data so that all numeric values are between 0-1. I use the min-max normalization which means that: <br>
$$x_{new} = \frac{x_{old}-x_{min}}{x_{max}-x_{min}}$$

### Data Binning
For the columns, `TotalBaseWaterVolume` and `TotalBaseNonWaterVolume`, I set up bins to better see the distribution of values. I then create a histogram to see the distribution.

```python
print(concat_df.dtypes) # get the data types of all the columns
concat_df.describe(include="all") # get a statistical summary of the data. include= 'all' gives summary of non-numeraical values too
```

```python
# cleaning the data: removing the columns that aren't needed
cleanUp = ['UploadKey','JobStartDate','JobEndDate','FederalWell',
           'IndianWell','IngredientKey', 'IngredientComment','ClaimantCompany','APINumber', 
           'StateNumber', 'CountyNumber', 'CASNumber','SystemApproach','IsWater','PurposePercentHFJob',
           'PurposeIngredientMSDS','DTMOD','Source','PurposeKey','FFVersion','DisclosureKey']
cleanData = concat_df.drop(labels=cleanUp, axis=1)
print('Removed unnecessary columns')

# remove the rows that don't have a value in TotalBaseWaterVolume and TotalBaseNonWaterVolume
cleanData.dropna(subset=['TotalBaseWaterVolume','TotalBaseNonWaterVolume'], inplace=True)
print(f'Removed rows that didn\'t have data in water/nonwater columns. There are {len(cleanData.index)} rows left')

# remove the duplicate rows based on all the following columns:
dup = ['OperatorName', 'WellName', 'Latitude','Longitude', 'Projection', 'TVD', 'TotalBaseWaterVolume',
       'TotalBaseNonWaterVolume', 'StateName', 'CountyName']
cleanData.drop_duplicates(subset=dup,inplace=True)
print(f'Duplicate rows have been removed. There are {len(cleanData.index)} rows left')

cleanData.describe(include='all')
```

```python
# Date Normalization on the following columns:
# TVD, PercentHighAdditive, PercentHFJob
import numpy as np
columns = ['TVD','PercentHighAdditive', 'PercentHFJob']

for col in columns:
    cleanData[col] = (cleanData[col]-cleanData[col].min())/(cleanData[col].max()-cleanData[col].min())
# Data Binning for TotalBaseWaterVolume, TotalBaseNonWaterVolume
# binNames = ['XSmall', 'Small', 'Medium', 'Large', 'XLarge']
binNames = [1, 2, 3, 4, 5]
for col in ['TotalBaseWaterVolume', 'TotalBaseNonWaterVolume']:
    print(f'{col=}\n\tmax is {max(cleanData[col])} and min is {min(cleanData[col])}')
    bins = np.linspace(min(cleanData[col]), max(cleanData[col]),6)
    print(f'\t{bins=}')
    dataBinned = col+'-Binned'
    cleanData[dataBinned] = pd.cut(cleanData[col], bins,labels=binNames, include_lowest=True)
    # convert to int
    cleanData[dataBinned] = cleanData[dataBinned].astype('int64')
print(cleanData.head(10))
cleanData.describe(include='all')




```

```python
# a histogram for the TotalBaseWaterVolume-Binned and TotalBaseNonWaterVolume-Binned columns
cleanData.hist(column= ['TotalBaseWaterVolume-Binned','TotalBaseNonWaterVolume-Binned'],bins=5)
```

## Grouping the data
Now that the data has been cleaned and processed, I group the data based on the `StateName` and `CountyName`. This allows me to calculate the total amount of water used in each county and each state.

```python
# create new DataFrame that includes TotalBaseWaterVolume, TotalBaseNonWaterVolume, StateName and CountyName
# cutLabels = ['TotalBaseWaterVolume', 'TotalBaseNonWaterVolume', 'StateName', 'CountyName']
cutDf = cleanData.loc[:,'TotalBaseWaterVolume':'CountyName']
print(cutDf.head(5))

# grouping the wells based on StateName and CountyName
groupLabels = ['StateName','CountyName']
groups = cutDf.groupby(groupLabels, group_keys=False, as_index=False)
# countyGroups = cutDf.groupby('CountyName', group_keys=False,as_index=False)
stateGroups = cutDf.groupby('StateName', group_keys=False,as_index=False)


# sum up the water columns for county
# countySum = countyGroups.sum()
# print(countySum.head(5))

# sum up the counties per state
groupedWaterSums = groups['TotalBaseWaterVolume'].sum()
groupedNonWaterSums = groups['TotalBaseNonWaterVolume'].sum()
print(groupedWaterSums.head())
print(groupedNonWaterSums.head())

# sum up the water columns for state
stateSum = stateGroups.sum()
print(stateSum)
```

```python
import matplotlib.pyplot as plt

# plot all states water and nonwater
states = list(stateSum['StateName'])
for state in states:
    countiesWater = groupedWaterSums[groupedWaterSums['StateName']==state].loc[:,'CountyName':]
    countiesNonWater = groupedNonWaterSums[groupedNonWaterSums['StateName']==state].loc[:,'CountyName':]
    # print(countiesWater,countiesNonWater,sep='\n')

    xLabels = list(countiesWater['CountyName'])
    waterVal = list(countiesWater['TotalBaseWaterVolume'])
    nonWaterVal = list(countiesNonWater['TotalBaseNonWaterVolume'])
    xAxis = np.arange(len(xLabels))
    #print(xLabels)
    # change the xLabels to just include the name, not 'State Waters -' or 'Offshore Islands'
    for idx,x in enumerate(xLabels):
        temp = x
        if 'State Waters -' in x:
            #print('Cutting out State Waters -')
            temp = x.replace('State Waters -','')
        if 'Offshore Islands' in x:
            #print('Cutting out Offshore Islands')
            temp = temp.replace('Offshore Islands','')
        if 'Quadrangle' in x:
            temp = temp.replace('Quadrangle', '')
        if 'West' in x:
            temp = temp.replace('West', 'W.')
        if 'East' in x:
            temp = temp.replace('East', 'E.')
        if 'North' in x:
            temp = temp.replace('North', 'N.')
        if 'South' in x:
            temp = temp.replace('South', 'S.')
        xLabels[idx] = temp
    #print(f'After trimming: \n{xLabels}')
    #print(len(xLabels))
    # changing the figsize based on how many wells the state has
    if len(xLabels) > 60:
        #print('Changing figsize width to 55 and height to 20')
        f = plt.figure()
        f.set_figwidth(55)
        f.set_figheight(20)
    elif len(xLabels) > 25:
        #print('Changing figsize width to 34 and height to 15')
        f = plt.figure()
        f.set_figwidth(34)
        f.set_figheight(15)
    elif len(xLabels) > 10:
        #print('Changing figsize width to 23')
        f = plt.figure()
        f.set_figwidth(23)
        f.set_figheight(10)
    elif len(xLabels) > 7:
        #print('Changing figsize width to 10')
        f = plt.figure()
        f.set_figwidth(10)
        f.set_figheight(10)
    # if none of the counties used NonWater material, don't include the nonwater in the chart (makes confusing)
    if all(v==0 for v in nonWaterVal):
        plt.bar(xAxis, waterVal, 0.4, label = 'Total Water Volume')
        plt.title("Volume of Water Used")
        print(f'Counties in the state of {state} did not use any nonwater materials for fracking')
    else:
        plt.bar(xAxis-0.2, waterVal, 0.4, label = 'Total Water Volume')
        plt.bar(xAxis + 0.2, nonWaterVal, 0.4, label = 'Total NonWater Volume')
        plt.title("Volume of Water and Nonwater Material Used")
        # for large numbers of counties, increase the size of the graph
    plt.xticks(xAxis, xLabels)
    plt.xlabel(state)
    # print(plt.yticks())
    # set the yticks to make chart easier to understand
    loc,labels = plt.yticks()
    for i in range(len(loc)):
        labels[i] = str(loc[i])[:-2] # no decimal points
    plt.yticks(loc,labels)
    plt.ylabel("Amount used (in gallons)")

    plt.legend()
    plt.show()
```

```python
# plotting the stateSum in a bar chart

# get the labels, water and nonwater values for the chart
xLabels = list(stateSum['StateName'])
waterVal = list(stateSum['TotalBaseWaterVolume'])
nonWaterVal = list(stateSum['TotalBaseNonWaterVolume'])

xAxis = np.arange(len(xLabels))

for idx,x in enumerate(xLabels):
    temp = x
    if 'West ' in x:
        temp = temp.replace('West', 'W.')
    if 'East ' in x:
        temp = temp.replace('East', 'E.')
    if 'North ' in x:
        temp = temp.replace('North', 'N.')
    if 'South ' in x:
        temp = temp.replace('South', 'S.')
    xLabels[idx] = temp

# start plotting the plots on a graph
# plt.figure(figsize=(70.0,40.0))
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(30)
plt.bar(xAxis, waterVal, 0.4, label = 'TotalBaseWaterVolume')
#plt.bar(xAxis + 0.2, nonWaterVal, 0.4, label = 'TotalBaseNonWaterVolume')

  
plt.xticks(xAxis, xLabels)
plt.xlabel("State")
plt.ylabel("Amount used (in gallons)")
# set the yticks to make chart easier to understand
loc,labels = plt.yticks()
for i in range(len(loc)):
    labels[i] = str(loc[i])[:-2] # no decimal points
plt.yticks(loc,labels)
plt.title("Volume of Water and Nonwater Material Used")
plt.legend()

plt.show()
```
